```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df= pd.read_excel(r"C:\Users\Vivek Pirsali\Desktop\salesonadv.xlsx")
df
```

```python
%matplotlib inline
plt.xlabel('Advertising($)')
plt.ylabel('sales($)')
plt.scatter(df.adv,df.sales,marker='*', color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['adv']],df.sales)
```

```python
reg.predict([[200]])
```

```python
%matplotlib inline
plt.xlabel('Advertising($)')
plt.ylabel('sales($)')
plt.scatter(df.adv,df.sales,marker='*', color='red')
plt.plot(df.adv,reg.predict(df[['adv']]),color='blue')
```
